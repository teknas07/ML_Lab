```
def sigmoid(z):
    if z.all() < 0 :
        return np.exp(z)/(1 + np.exp(z))
    else:
        return 1 / (1 + np.exp(-z))

V = np.random.randint(-1000,1000,(10))
v = []
for x in V:
    v.append(sigmoid(x))
print(v)
print(V)
```

```
def hypothesis(theta, X):
    if sigmoid(np.dot(theta, X)) >= 0.5:
        print(0)
    else:
        print(1)
```

```
def cost(theta, X, y):
    return(-y * np.log(hypothesis(theta, X)) - (1-y) * np.log(1-hypothesis(theta, X))).mean()
```
